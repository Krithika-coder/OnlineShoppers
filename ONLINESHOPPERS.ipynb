```python
x=data['BounceRates'].values.reshape(-1,1)
y=data['ExitRates'].values.reshape(-1,1)
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

#LINEAR REGRESSION AND SCI-KIT

```python
# Logistic Regression 3-class Classifier
import numpy as np
from sklearn.linear_model import LogisticRegression
# Splitting the dataset into the Training set and Test set
X =data['Region'].values.reshape(-1,1)
Y =data['TrafficType'].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
# Fitting Logistic Regression to the Training set :
classifier = LogisticRegression(random_state = 0, solver='lbfgs', multi_class='auto')
classifier.fit(X_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(X_test)
# Predict probabilities
probs_y=classifier.predict_proba(X_test)
```

```python
#Making the Confusion Matrix :
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Plot confusion matrix
import seaborn as sns
import pandas as pd
# confusion matrix sns heatmap 
ax = plt.axes()
z =cm
sns.heatmap(z, annot=True, annot_kws={"size": 20}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix')
plt.show()
```
